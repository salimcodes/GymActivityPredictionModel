## Model training

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score, precision_score, recall_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

# Print the evaluation metrics
print("Decisoion Tree Classifier")
print(f"F1-Score: {f1}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Accuracy: {accuracy}")

# Calculate and plot the confusion matrix as an image
confusion_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion_mat,
        annot=True,
        fmt='g')
plt.ylabel('Prediction',fontsize=14)
plt.xlabel('Actual',fontsize=14)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

# Print the evaluation metrics
print("Random Forest Classifer Model")
print(f"F1-Score: {f1}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Accuracy: {accuracy}")

# Calculate and plot the confusion matrix as an image
confusion_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion_mat,
            annot=True,
            fmt='g')
plt.ylabel('Prediction',fontsize=14)
plt.xlabel('Actual',fontsize=14)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```

## Model training with Sliding Window

```python
X_windows = []
y_windows = []
window_size = 100
stride = 10
for i in range(0, len(X) - window_size + 1, stride):
    X_window = pd.DataFrame(X[i:i+window_size])
    y_window = pd.Series(y[i:i+window_size])
    X_windows.append(X_window)
    y_windows.append(y_window)
        
X_windows = pd.concat(X_windows)
y_windows = pd.concat(y_windows)
X_train_slid, X_test_slid, y_train, y_test = train_test_split(X_windows, y_windows, test_size=0.2, random_state=42)

model = DecisionTreeClassifier()
model.fit(X_train_slid, y_train)

y_pred = model.predict(X_test_slid)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')

# Print the evaluation metrics
print("Decision Tree Classifier with Sliding Window")
print(f"F1-Score: {f1}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Accuracy: {accuracy}")
# Calculate and plot the confusion matrix as an image
confusion_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion_mat,
            annot=True,
            fmt='g')
plt.ylabel('Prediction',fontsize=14)
plt.xlabel('Actual',fontsize=14)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```
